```python
import os

input_path = r"C:\Users\kuzey\OneDrive\Masa√ºst√º\Curley's Wife\data\Of Mice and Men-John Steinbeck-Text.txt"
output_dir = r"C:\Users\kuzey\OneDrive\Masa√ºst√º\Curley's Wife\data\sections_txt"
chunk_size_words = 1000

with open(input_path, "r", encoding="utf-8") as f:
    full_text = f.read()

words = full_text.split()
chunks = [" ".join(words[i:i + chunk_size_words]) for i in range(0, len(words), chunk_size_words)]

os.makedirs(output_dir, exist_ok=True)

for i, chunk in enumerate(chunks):
    file_path = os.path.join(output_dir, f"chunk_{i+1:03}.txt")
    with open(file_path, "w", encoding="utf-8") as f:
        f.write(chunk)

print(f"‚úÖ Done! Saved {len(chunks)} chunks to: {output_dir}")
```

```python
import pandas as pd

csv_path = r"C:\Users\kuzey\OneDrive\Masa√ºst√º\Curley's Wife\data\curleys_quotes.csv"
df = pd.read_csv(csv_path)

def safe_fix(text):
    if isinstance(text, str):
        try:
            return text.encode('cp1252').decode('utf-8')
        except UnicodeDecodeError:
            return text  
    return text

for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].apply(safe_fix)

cleaned_path = csv_path.replace(".csv", "_cleaned.csv")
df.to_csv(cleaned_path, index=False, encoding="utf-8")

print(f"‚úÖ Cleaned and saved to: {cleaned_path}")
```

```python
import pandas as pd

csv_path = r"C:\Users\kuzey\OneDrive\Masa√ºst√º\Curley's Wife\data\curleys_quotes_cleaned.csv"
df = pd.read_csv(csv_path)

REPLACEMENTS = {
    '√¢‚Ç¨‚Ñ¢': '‚Äô',
    '√¢‚Ç¨‚Äú': '‚Äì',
    '√¢‚Ç¨≈ì': '‚Äú',
    '√¢‚Ç¨¬ù': '‚Äù',
    '√¢‚Ç¨Àú': '‚Äò',
    '√¢‚Ç¨¬¶': '‚Ä¶',
    '√É¬©': '√©',
    '√É¬®': '√®',
    '√É': '√†',
    '√Ç': '', 
}

def clean_text(text):
    if not isinstance(text, str):
        return text
    for broken, fixed in REPLACEMENTS.items():
        text = text.replace(broken, fixed)
    return text

for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].apply(clean_text)

cleaned_path = csv_path.replace(".csv", "_even_more_cleaned.csv")
df.to_csv(cleaned_path, index=False, encoding="utf-8")

print(f"‚úÖ Final cleaned file saved to:\n{cleaned_path}")
```
